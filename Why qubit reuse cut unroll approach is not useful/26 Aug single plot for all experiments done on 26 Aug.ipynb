```python
import json

with open("25 Aug whether cut unroll combined then cut is useful.json", "r") as jsonfile:
    data = json.load(jsonfile)
```

```python
data
```

```python
experiments=["experiment 1", "experiment 2","experiment 3","experiment 4","experiment 5","experiment 6"]
Y= [data[experiments[i]]['obtained expectation value'] for i in range(6)]
```

```python
import matplotlib.pyplot as plt


x=[1,2,3,4,5,6]
# Create bar plot
for i in range(len(Y)):
    if Y[i]<0:
        Y[i]=-Y[i]
        
plt.bar(x, Y)

# Add title and labels
plt.title('7 qubit GHZ state weight 1 Z observables expectation value')
plt.xlabel('Experiments')
plt.ylabel('Expectation value')

# Show plot
plt.show()
```
